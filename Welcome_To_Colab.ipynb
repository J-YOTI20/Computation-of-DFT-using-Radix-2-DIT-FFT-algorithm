```
# tep_mlp_demo.py
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay

# --------------------------------------------------
# 1) LOAD DATA (synthetic if CSVs missing)
# --------------------------------------------------

try:
    df0 = pd.read_csv("fault_free_training.csv")
    dfF = pd.read_csv("faulty_training.csv")
    df = pd.concat([df0, dfF], ignore_index=True)
    print("Loaded real TEP dataset.")
except:
    print("CSV files not found. Generating synthetic TEP-like dataset...")
    n_samples_per_class = 300
    n_vars = 52
    n_classes = 6  # e.g., normal + 5 faults
    rows = []
    for cls in range(n_classes):
        for i in range(n_samples_per_class):
            base = np.random.randn(n_vars) + cls*0.5
            rows.append([cls] + list(base))
    cols = ["faultNumber"] + [f"PV_{i+1}" for i in range(n_vars)]
    df = pd.DataFrame(rows, columns=cols)

# --------------------------------------------------
# 2) PREPROCESSING
# --------------------------------------------------

feature_cols = [c for c in df.columns if c != "faultNumber"]
X = df[feature_cols].values
y = df["faultNumber"].values

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# --------------------------------------------------
# 3) PCA VISUALIZATION
# --------------------------------------------------

pca = PCA(n_components=2, random_state=42)
X_pca = pca.fit_transform(X_scaled)

plt.figure(figsize=(8,6))
for cls in np.unique(y):
    mask = (y == cls)
    plt.scatter(X_pca[mask,0], X_pca[mask,1], label=f"Class {cls}", s=18)
plt.title("PCA Scatter Plot — Synthetic TEP")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.legend()
plt.tight_layout()
plt.show()

# --------------------------------------------------
# 4) TRAIN-TEST SPLIT
# --------------------------------------------------

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.25, random_state=42, stratify=y
)

# --------------------------------------------------
# 5) MLP NEURAL NETWORK
# --------------------------------------------------

mlp = MLPClassifier(hidden_layer_sizes=(256,128),
                    activation="relu",
                    solver="adam",
                    max_iter=300,
                    early_stopping=True,
                    random_state=42)
print("Training MLP...")
mlp.fit(X_train, y_train)

# Training loss curve
plt.figure(figsize=(6,4))
plt.plot(mlp.loss_curve_)
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("MLP Training Loss Curve")
plt.show()

# --------------------------------------------------
# 6) EVALUATION
# --------------------------------------------------

y_pred = mlp.predict(X_test)
print("=== MLP Classification Report ===")
print(classification_report(y_test, y_pred, zero_division=0))

# Confusion matrix
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot(cmap="Blues", xticks_rotation=90)
plt.title("MLP Confusion Matrix")
plt.show()

# --------------------------------------------------
# 7) SVM DECISION BOUNDARY (PCA 2D)
# --------------------------------------------------

svm = SVC(kernel="rbf", gamma="scale")
svm.fit(X_pca, y)

x_min, x_max = X_pca[:,0].min()-1, X_pca[:,0].max()+1
y_min, y_max = X_pca[:,1].min()-1, X_pca[:,1].max()+1
xx, yy = np.meshgrid(np.linspace(x_min,x_max,300), np.linspace(y_min,y_max,300))
Z = svm.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)

plt.figure(figsize=(8,6))
plt.contourf(xx, yy, Z, alpha=0.3, cmap="tab10")
plt.scatter(X_pca[:,0], X_pca[:,1], c=y, s=12, cmap="tab10")
plt.title("SVM Decision Boundary (PCA 2D)")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
```

```
# -------------------------
# Nanoelectronics: Process Optimisation and Fault Detection
# Full ML pipeline with PCA and MLP Neural Network
# -------------------------

import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay

# -------------------------
# 1. Data Loading
# -------------------------
try:
    df_normal = pd.read_csv("fault_free_training.csv")
    df_faulty = pd.read_csv("faulty_training.csv")
    df = pd.concat([df_normal, df_faulty], axis=0)
    print("Loaded real TEP dataset.")
except FileNotFoundError:
    print("Real dataset not found. Creating synthetic dataset for demonstration.")
    n_samples = 3000
    n_vars = 52
    X = np.random.randn(n_samples, n_vars)
    y = np.random.randint(0, 6, n_samples)  # 6 classes demo
    df = pd.DataFrame(X, columns=[f"var{i+1}" for i in range(n_vars)])
    df["faultNumber"] = y

df.head()

# -------------------------
# 2. Preprocessing
# -------------------------
feature_cols = [c for c in df.columns if c != "faultNumber"]
X = df[feature_cols].values
y = df["faultNumber"].values

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# -------------------------
# 3. PCA Visualization
# -------------------------
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

plt.figure(figsize=(8,6))
scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='tab20', s=15)
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("PCA Scatter Plot (2 Components)")
plt.colorbar(scatter, label="Fault Class")
plt.show()

# -------------------------
# 4. Train-Test Split
# -------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.25, random_state=42, stratify=y
)

# -------------------------
# 5. MLP Neural Network
# -------------------------
from sklearn.neural_network import MLPClassifier

mlp = MLPClassifier(
    hidden_layer_sizes=(128, 64),
    activation='relu',
    solver='adam',
    max_iter=100,
    random_state=42
)

mlp.fit(X_train, y_train)
y_pred = mlp.predict(X_test)

# -------------------------
# 6. Evaluation
# -------------------------
print("Classification Report:")
print(classification_report(y_test, y_pred, zero_division=0))

cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title("Confusion Matrix - MLP Neural Network")
plt.xlabel("Predicted Class")
plt.ylabel("True Class")
plt.show()

# -------------------------
# 7. Correlation Matrix
# -------------------------
# Limit to first 30 variables for readability
subset_cols = feature_cols[:30]
corr = df[subset_cols].corr()

plt.figure(figsize=(10,8))
sns.heatmap(corr, cmap='coolwarm', center=0, annot=False)
plt.title("Correlation Matrix (First 30 Variables)")
plt.show()
```
